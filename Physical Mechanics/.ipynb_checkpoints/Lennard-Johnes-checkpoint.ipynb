```python
import numpy as np
from sympy import symbols
import scipy.integrate as integrate
import scipy.special as special
from matplotlib import pyplot as plt
```

```python
def U(r):
    return U0 * ((d0/r)**12 - (d0/r)**6)

def B(T):
    return 2*np.pi*integrate.quad(lambda r: (1-np.exp(-U(r)/T))*r**2, 0, 50*d0)[0]

U0 = 200
d0 = 2*10**(-10)

N = 100
t = np.linspace(10, 1500, num = N)
arr = []

for i in range(N):
    arr.append(B(t[i]))
    
plt.figure(figsize = (10, 5))
plt.plot(t, arr, '.', color = 'crimson');
plt.xlabel('T, K');
plt.ylabel('B(T)');
plt.grid()
```
