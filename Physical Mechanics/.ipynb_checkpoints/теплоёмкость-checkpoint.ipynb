```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

def z_(T):
    l = np.arange(100000)
    el = (2*l+1)*np.exp(1)**(-3*l*(l+1)/T)
    return float(nsum(lambda l: (2*l+1)*np.exp(1)**(-3*l*(l+1)/T), [0, inf]))

T = 1000

t = np.linspace(T, T*10, 100)

def dz_(T):
    l = np.arange(50)
    el = (2*l+1)*np.exp(1)**(-3*l*(l+1)/T)*3*l*(l+1)/T**2
    return float(nsum(lambda l: (2*l+1)*np.exp(1)**(-3*l*(l+1)/T)*3*l*(l+1)/T**2, [0, inf]))

def u(T):
    el = T**2/z_(T)
    return el.sum()

def du(T):
    l = np.arange(50)
    el = (2*T*z_(T) - dz_(T)*T**2)/z_(T)**2
    return el

def dv(T):
    l = np.arange(50)
    el = (2*l+1)*np.exp(1)**(-3*l*(l+1)/T)*3*l*(l+1)/T**2 * (3*l*(l+1)/T**2 - 2/T)
    return float(nsum(lambda l: (2*l+1)*np.exp(1)**(-3*l*(l+1)/T)*3*l*(l+1)/T**2 * (3*l*(l+1)/T**2 - 2/T), [0, inf]))

def sum_el(T):
    l = np.arange(50)
    return u(T)*dv(T) + du(T)*dz_(T)

p = []
for i in range(len(t)):
    print(i)
    p.append(sum_el(t[i]))

plt.plot(t, p)
plt.show()
```

Здесь хорошо виден *effect*: теплоёмкость достигает максимума.

```python
T = 100

t = np.linspace(0.5, T/2, 400)

p2 = []
for i in range(len(t)):
    p2.append(sum_el(t[i]))

plt.figure(figsize = (10, 5))
plt.plot(t, p2)
plt.xlabel('T')
plt.ylabel('c')
plt.title('Температурная зависимость теплоёмкости')
plt.show()
```

```python
print('Точка максимума - около ' + str(round(np.linspace(0.5, T/2, 400)[np.argmax(p2)], 2)) +' К')
```
